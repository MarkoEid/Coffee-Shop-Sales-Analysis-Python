# fixing the time to be more readable

```python
df["before_semicolon"] = df["Time"].str.split(":").str[0:2]
```

```python
df
```

```python
def x(l):
    s = l[0]+ ":" + l[1]
    return s
```

```python
df['Time2'] = df['before_semicolon'].apply(x)
```

```python
df.head()
```

```python
df.drop(columns=["before_semicolon"],inplace=True)
```

```python
df.drop(columns=["Time"],inplace=True)
```

```python
df.rename(columns={"Time2" : "Time"}, inplace=True)
```

# check at what month they earn a lot

```python
x = df.pivot_table(
    index=pd.Grouper(freq='M', key='Date'),
    values='money',      
    aggfunc='sum'
)
x
```

```python
x.describe()
```

```python
with pd.ExcelWriter('stocks_weather.xlsx') as writer:
    x.to_excel(writer, sheet_name="Pro-Month",index=False)
    Time_money.to_excel(writer, sheet_name="Tim_mon",index=False)
    TimeDay_money.to_excel(writer, sheet_name="Tim_Day_mon",index=False)
    Caffe_Category.to_excel(writer, sheet_name="Caff_categ",index=False)
```

# Coffee Shop Sales Summary

- Latte is the most purchased item, accounting for 24% of total sales.
  
- There is only one payment type, which is card.

- The afternoon is the time of day when the café earns the most, with 34% of total sales.

- All items have increased by 5.88%.

- In March, the café’s earnings increased by 14%, while in April, they decreased by 5%.

- On Tuesday, earnings were the highest at 16%, while on Sunday, they were the lowest at 11%.

- During the afternoon, morning, and night, the café earns roughly 34% in each period.

- At 10:12, there are a lot of customer requests.

- From 2025-02-28 to 2025-03-31, earnings increased by 11%, and from 2024-10-31 to 2024-11-30, by 12%.
